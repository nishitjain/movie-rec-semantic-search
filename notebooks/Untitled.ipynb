```python
movies = pd.read_csv(DATA_PATH+"movies_metadata.csv")
```

```python
def parse_dict(d, key):
    try:
        name = eval(d)[key]
        return name
    except:
        return ""

def parse_lst_dict(lst, key):
    names = []
    try:
        for d in eval(lst):
            names.append(d[key])
    except:
        pass
    return names
```
